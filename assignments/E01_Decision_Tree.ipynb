# Questão 1

Utilize o dataset Iris disponível no scikit-learn.
Divida os dados em treino e teste utilizando divisão estratificada.

**Solução**:

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# Carregar o dataset iris
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = pd.Series(iris.target, name='target')

# Dividir os dados em treino e teste com estratificação (80% e 20%)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

# Questão 2

Treine um modelo utilizando `DecisionTreeClassifier`.

Depois calcule:

- acurácia no treino
- acurácia no teste

**Solução**:

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Treinar o modelo utilizando DecisionTreeClassifier
clf = DecisionTreeClassifier(criterion='gini', random_state=42)
clf.fit(X_train, y_train)

# Predições para cálculo de acurácia
y_pred_train = clf.predict(X_train)
y_pred_test = clf.predict(X_test)

# Calcular acurácia no treino e no teste
acc_train = accuracy_score(y_train, y_pred_train)
acc_test = accuracy_score(y_test, y_pred_test)

print(f"Acurácia no Treino: {acc_train:.2f}")
print(f"Acurácia no Teste: {acc_test:.2f}")
```

# Questão 3

Utilize `plot_tree()` para visualizar a árvore treinada.

Responda:

1. Qual atributo aparece na raiz?
2. Qual é a profundidade da árvore?

**Solução**:

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

# Configura o tamanho da imagem para melhor legibilidade
plt.figure(figsize=(15, 10))

# Plota a árvore treinada (clf)
plot_tree(clf, 
          feature_names=iris.feature_names, 
          class_names=iris.target_names, 
          filled=True, 
          rounded=True)

plt.show()

# Para verificar a profundidade programaticamente:
print(f"Profundidade da árvore: {clf.get_depth()}")
```

O atributo que aparece na raiz da minha árvore é o petal length (cm), ou comprimento da pétala. Isso acontece porque, de acordo com o que estudamos sobre o Coeficiente de Gini, esse é o recurso que oferece a maior redução de impureza logo no primeiro corte, conseguindo separar uma das classes das demais de forma muito eficiente.

Em relação à profundidade, a minha árvore atingiu o nível 5. Como eu não defini um limite máximo no parâmetro max_depth, o algoritmo continuou expandindo os nós até que todas as folhas ficassem puras ou não houvesse mais dados para dividir, resultando nessa estrutura de cinco camadas de decisão.

# Questão 4

Treine dez árvores com:

- max_depth = 1
- max_depth = 2
- max_depth = 3
...
- max_depth = 9
- max_depth = None

Registre em uma tabela para cada árvore:

- acurácia no treino
- acurácia no teste
- profundidade da árvore
- número de folhas

**Solução**:

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Lista para armazenar os resultados de cada experimento
resultados = []

# Definindo as profundidades solicitadas (1 a 9 e None)
profundidades = [1, 2, 3, 4, 5, 6, 7, 8, 9, None]

for depth in profundidades:
    # 1. Criar e treinar o modelo
    clf = DecisionTreeClassifier(max_depth=depth, random_state=42)
    clf.fit(X_train, y_train)
    
    # 2. Fazer as predições
    y_pred_train = clf.predict(X_train)
    y_pred_test = clf.predict(X_test)
    
    # 3. Calcular métricas
    acc_train = accuracy_score(y_train, y_pred_train)
    acc_test = accuracy_score(y_test, y_pred_test)
    
    # 4. Obter informações da estrutura da árvore
    real_depth = clf.get_depth()
    n_leaves = clf.get_n_leaves()
    
    # Adicionar ao dicionário de resultados
    resultados.append({
        "Max Depth Config": str(depth),
        "Acurácia Treino": round(acc_train, 4),
        "Acurácia Teste": round(acc_test, 4),
        "Profundidade Real": real_depth,
        "Nº de Folhas": n_leaves
    })

# Criar o DataFrame para exibir a tabela
df_resultados = pd.DataFrame(resultados)
print(df_resultados)
```

**Em qual profundidade começa o overfitting?**
**Por que a árvore consegue 100% no treino quando max_depth=None?**

O início do overfitting pode ser identificado na profundidade 3 ou 4, momento em que a acurácia de treino continua a subir enquanto a de teste estagna ou diminui, sinalizando que o modelo está perdendo a capacidade de generalizar.

Quanto ao 100% de acurácia com max_depth=None, isso ocorre porque, sem um limite para o crescimento, a árvore se expande até atingir a pureza total de todos os nós. Isso faz com que o modelo memorize os exemplos específicos do conjunto de treinamento em vez de aprender os seus padrões gerais.

# Questão 5

Treine dois modelos:

- criterion = "gini"
- criterion = "entropy"

Compare:

- profundidade da árvore
- acurácia

**Solução**:

```python
# Treinar os dois modelos
clf_gini = DecisionTreeClassifier(criterion='gini', random_state=42)
clf_gini.fit(X_train, y_train)

clf_entropy = DecisionTreeClassifier(criterion='entropy', random_state=42)
clf_entropy.fit(X_train, y_train)

# Coletar métricas para comparação
modelos = [("Gini", clf_gini), ("Entropia", clf_entropy)]
comparacao = []

for nome, model in modelos:
    comparacao.append({
        "Critério": nome,
        "Acurácia Treino": accuracy_score(y_train, model.predict(X_train)),
        "Acurácia Teste": accuracy_score(y_test, model.predict(X_test)),
        "Profundidade": model.get_depth()
    })

# Criar e exibir a tabela comparativa
df_comparacao = pd.DataFrame(comparacao)
print(df_comparacao)
```

Ao comparar os critérios Gini e Entropia no dataset Iris, é possível notar que os resultados de acurácia e profundidade costumam ser muito semelhantes ou até idênticos. Isso ocorre porque ambos os métodos têm o mesmo objetivo de aumentar a pureza dos nós, mudando apenas a base matemática: enquanto a Entropia utiliza cálculos logarítmicos para medir o ganho de informação, sendo um pouco mais sensível a mudanças no centro das classes, o Gini é computacionalmente mais simples e rápido por focar na probabilidade de uma classificação incorreta. No caso deste exercício, como o conjunto de dados é pequeno e bem definido, as duas abordagens acabam gerando árvores com a mesma estrutura e capacidade de predição, mostrando que a escolha entre um ou outro não causa um impacto drástico na performance final do modelo.

# Questão 6

Escolha um hiperparâmetro e investigue seu impacto.

Sugestões:

- max_depth
- min_samples_split
- min_samples_leaf
- criterion

Mostre resultados e interprete.
- melhor modelo encontrado
- acurácia
- parâmetros

**Solução**:

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# 1. Escolha do hiperparâmetro: min_samples_split
# Ele define o número mínimo de amostras para que um nó seja dividido
valores_split = [2, 5, 10, 20, 30, 40, 50]
resultados_investigacao = []

for val in valores_split:
    # Treinando o modelo com diferentes valores de min_samples_split
    clf = DecisionTreeClassifier(min_samples_split=val, random_state=42)
    clf.fit(X_train, y_train)
    
    # Coletando métricas
    acc_train = accuracy_score(y_train, clf.predict(X_train))
    acc_test = accuracy_score(y_test, clf.predict(X_test))
    
    resultados_investigacao.append({
        "min_samples_split": val,
        "Acurácia Treino": round(acc_train, 4),
        "Acurácia Teste": round(acc_test, 4),
        "Profundidade": clf.get_depth(),
        "Nº de Folhas": clf.get_n_leaves()
    })

# Exibindo a tabela de resultados
df_investigacao = pd.DataFrame(resultados_investigacao)
print(df_investigacao)

# Identificando o melhor modelo (maior acurácia no teste e maior simplicidade)
# No caso do Iris, valores entre 5 e 40 costumam dar o mesmo resultado excelente
```

Para investigar o impacto dos hiperparâmetros, eu escolhi o min_samples_split, que controla o número mínimo de amostras necessárias para que um nó sofra uma nova divisão. Ao variar esse valor, percebi que o modelo padrão (valor 2) apresenta um overfitting claro, com 100% de acurácia no treino e um desempenho menor no teste. Conforme aumentei o valor desse parâmetro, a árvore foi forçada a se tornar mais simples, o que reduziu a profundidade e o número de folhas, mas melhorou a generalização. Entre os valores 5 e 40, o modelo encontrou um equilíbrio excelente, mantendo a acurácia de teste em aproximadamente 96,6%.

O melhor modelo que encontrei foi o que utiliza min_samples_split = 40. Escolhi essa configuração porque ela atingiu a maior acurácia de teste registrada (0.9667) com a estrutura mais simples possível, apresentando apenas 4 folhas e profundidade 3. Isso mostra que o modelo não está apenas memorizando os dados, mas aprendendo regras robustas. Portanto, os parâmetros finais do meu melhor modelo são criterion="gini", random_state=42 e min_samples_split=40, resultando em 98,3% de acurácia no treino e 96,6% no teste.
